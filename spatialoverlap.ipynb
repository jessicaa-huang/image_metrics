```python
import spatialoverlap as S
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
spatial_overlap_map = {}
satellite_dir = "University-Release/train/satellite"

for dataset in os.listdir(satellite_dir):
    folder_path = f"University-Release/train/drone/{dataset}"
    ref_image = cv2.imread(f"{satellite_dir}/{dataset}/{dataset}.jpg")
    res = S.spatial_overlap(folder_path, ref_image)
    spatial_overlap_map[dataset] = res
    # print(f"Spatial overlap for {dataset}: {res}")

    plt.figure(figsize=(8, 6))
    plt.hist(res)
    plt.title(f'Satellite vs. Drone Images ({dataset})')
    plt.xlabel('Spatial Overlap')
    plt.ylabel('Frequency')
    plt.switch_backend('Agg')

    output_folder = "plots/spatialoverlap"
    output_path = os.path.join(output_folder, f"{dataset}_spatialoverlap.png")
    plt.savefig(output_path, dpi=300)
    print(f"Plot saved to {output_path}")
    plt.close()
```

```python
for key, result in spatial_overlap_map.items():
    print(f"{key}: {result}")
```
