```python
import spatialoverlap as S
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("University-Release/train/drone/0839/image-25.jpeg")
S.compute_feature_density(img)
```

```python
feature_density_map = {}
feature_density_only = []
dataset = "0842"
maindir = f"University-Release/train/drone/{dataset}"

for image in os.listdir(maindir):
    full_path = f"{maindir}/{image}"
    readimg = cv2.imread(full_path)
    res = S.compute_feature_density(readimg)
    feature_density_map[dataset] = res
    feature_density_only.append(res)
print(feature_density_only)
```

```python
feature_density_map = {}
feature_density_only = []
maindir = f"University-Release/train/satellite"

for dataset in os.listdir(maindir):
    full_path = f"{maindir}/{dataset}/{dataset}.jpg"
    readimg = cv2.imread(full_path)
    res = S.compute_feature_density(readimg)
    feature_density_map[dataset] = res
    feature_density_only.append(res)
print(feature_density_only)
```
